```python
from pyspark import SparkContext
try:
    sc.stop()
except:
    pass
from pyspark import SparkConf
sc = SparkContext.getOrCreate(
    SparkConf().setMaster("local[*]").set("spark.executor.memory", "4g").set("spark.driver.memory", "4g"))
sc.setLogLevel('WARN')
input_file="/content/drive/MyDrive/data/sample_data.csv"


ub_sample = sc.textFile(input_file)
header = ub_sample.first()
def covert_to_list(a):
  return [a]
def covert_extend(a, b):
  a.extend(b)
  return a
def append(a, b):
  a.append(b)
  return a
threshold=7
ub_sample = ub_sample.filter(lambda x: x != header).map(lambda x: tuple(x.split(","))).distinct().persist()

filtered_uid_bid = ub_sample.combineByKey(covert_to_list,append , covert_extend).map(lambda x: (x[0], list(set(x[1])))).filter(
    lambda x: len(list(set(x[1]))) >= threshold).collectAsMap()

uid_pairs = list(combinations(sorted(filtered_uid_bid.keys()), 2))

```

```python
edges = []
vertex_list = []

for i in uid_pairs:
    user1_list = filtered_uid_bid[i[0]]
    user2_list = filtered_uid_bid[i[1]]
    co_bid = list(set(user1_list) & set(user2_list))
    if len(co_bid) >= threshold:
        edges.append(i)
        edges.append((i[1], i[0]))
        vertex_list.extend([(i[0]), (i[1])])

vertex_list = list(set(vertex_list))
```

```python
#vertex: đỉnh
#edge: cạnh
def bfs(root, graph):
    queue = [root]#xét đỉnh ban đầu được chọn là được chọn là đỉnh gốc (đỉnh này ngẫu nhiên)
    visited = [root]
    level = {}
    parents = {}
    shortest_path = {}

    for vertex in graph.keys():#duyệt qua các đỉnh
        level[vertex] = -1 # set node 
        parents[vertex] = []#list các parent của mỗi đỉnh
        shortest_path[vertex] = 0# set đường đi ngắn nhất ban đầu của mỗi đỉnh là 0
    shortest_path[root] = 1 #set đường đi ngắn nhất của mỗi node =1 từ node gốc
    level[root] = 0
#dùng bfs theo queue
    while queue:
        current = queue.pop(0)# lấy phần tử đầu tiên
        visited.append(current)#check đã viếng thăm
        children = graph[current]#lấy đỉnh đầu tiên
        for child in children: #duyệt qua các node để biết lever của chúng
            if level[child] == -1: #nếu ban đầu lever chưa có thì thêm vào hàng đợi 
                queue.append(child)
                level[child] = level[current] + 1#tăng lever khi mà đã thêm vào hàng đợi
            if level[child] == level[current] + 1:# nếu đã có lever rồi
                parents[child].append(current)#thêm vào list các parent 
                shortest_path[child] = shortest_path[child] + shortest_path[current]#tính đường đi ngắn nhất từ root đến child

    return visited, parents, shortest_path
```

```python
def calculate_edge_credit(visited, parents, shortest_path):#hàm tính credit
    edge_credit = {}#khởi tạo credit cạnh
    vertex_credit = {vertex: 1 for vertex in visited} #set credit mỗi cạnh là 1
    for vertex in visited[::-1]:#duyệt qua các đỉnh trong list ghé thăm
        for parent in parents[vertex]:# duyệt qua các parent
            pair = min(vertex, parent), max(vertex, parent)
            credit = vertex_credit[vertex] * shortest_path[parent] / shortest_path[vertex]#credit sẽ bằng label*shorted_path_root to Y(i)/tổng số dduwognf đi ngắn nhất 
            if pair not in edge_credit.keys():
                edge_credit[pair] = 0# set các cặp cạnh chưa có thì set là 0
            edge_credit[pair] += credit#tính credit của các cặp đỉnh
            vertex_credit[parent] += credit# tính credit của parent= 1+ tổng credit của các con
    return edge_credit
```

```python
def calculate_betweenness(vertices, graph):
    betweenness = {}
    for vertex in vertices:
        visited, parents, shortest_path = bfs(vertex, graph)
        edge_credit = calculate_edge_credit(visited, parents, shortest_path)
        for edge, credit in edge_credit.items():
            if edge not in betweenness.keys():
                betweenness[edge] = 0# nếu cạnh chưa có bw ta set là 0
            betweenness[edge] += credit / 2# tính lại tất cả các bw , do mỗi node được tính 2  2 lần ta phải chia 2,
    return sorted(betweenness.items(), key=operator.itemgetter(1), reverse=True)
```

```python
def create_community(graph, vertex):#tạo com
    community = set()
    queue = [vertex]
    while queue:
        current = queue.pop(0)
        community.add(current)
        for neighbor in graph[current]:
            if neighbor not in community:#nếu chưa có thì 
                queue.append(neighbor)# thêm các phàn tử vào com
    return community
```

```python
def create_communities(graph, vertices):
    temp = vertices.copy()
    communities = []
    while temp:
        community = create_community(graph, temp.pop())
        communities.append(sorted(list(community)))
        temp = temp.difference(community)
    return communities
```

```python
def calculate_modularity(communities, graph, number_of_edges):
    total_modularity = 0.0
    for community in communities:
        for i in community:
            for j in community:
                k_i = len(graph[i]) # bậc của i 
                k_j = len(graph[j])# bậc của j
                a_ij = 1.0 if j in graph[i] else 0.0# 
                total_modularity += a_ij - k_i * k_j / (2 * number_of_edges) # mod=(1/2m)(Aij-Kj*Ki/2m)
    return total_modularity / (2 * number_of_edges)
```

```python
import copy
vertex_list = set(vertex_list)
def community_detection(vertices, graph, betweenness):
    number_of_edges = len(betweenness)
    btw, new_graph = betweenness.copy(), graph.copy()
    max_modularity, max_communities = -1, []
    while btw:
        communities = create_communities(new_graph, vertices)
        modularity = calculate_modularity(communities, graph, number_of_edges)
        if max_modularity < modularity:#tìm max modularity
            max_communities = communities
            max_modularity = modularity
        max_betweenness = max([v for _, v in btw])#tìm max modularity
        removed_edges = [edge for edge, value in btw if value ==max_betweenness ]#remove cạnh nào có max_modularity max_betweenness
        print("len edge",len(removed_edges))
        print(removed_edges)
        for edge in removed_edges:
            new_graph[edge[0]] = {i for i in new_graph[edge[0]] if i != edge[1]}
            #print("new graph",new_graph[edge[0]])
            new_graph[edge[1]] = {i for i in new_graph[edge[1]] if i != edge[0]}
            
        btw = calculate_betweenness(vertices, new_graph)
    return max_communities
```

```python
betweenness = calculate_betweenness(vertex_list, graph_actual)
result = community_detection(vertex_list,graph_actual, betweenness)
result.sort(key=lambda x: (len(x), x[0]))
```
